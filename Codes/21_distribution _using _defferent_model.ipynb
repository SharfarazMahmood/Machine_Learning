```
from sklearn.datasets  import load_iris
import matplotlib.pyplot as plt

iris = load_iris()
print(type(iris), "\n ",type(iris.data))
# print(iris.data)

n_sample, n_features = iris.data.shape
print(n_sample, n_features)
print(iris.target)
print(iris.target_names)


## slicing
print(iris.data[:10])
```

```
import numpy as np
from sklearn.neural_network import MLPClassifier
from sklearn.naive_bayes import GaussianNB

x = np.linspace(0,7,70)
# print(x)
y = np.linspace(0,3,30)
# print(y)

xx, yy = np.meshgrid(x,y)
# print(xx,"\n" , yy)
# plt.scatter(xx,yy)

print(xx.shape,"\n" , yy.shape)

l, w = xx.shape
n_test = l*w
xx = xx.reshape(n_test, 1)
yy = yy.reshape(n_test, 1)
print(xx.shape, yy.shape)

test_data = np.concatenate([xx, yy], axis=1)
print(test_data.shape)


classifier1 = MLPClassifier(max_iter=1000)
classifier1.fit( iris.data[:, 2:4] , iris.target)
print(classifier1.score)
predicted1 = classifier1.predict(test_data)
print(predicted1)

classifier2 = GaussianNB()
classifier2.fit( iris.data[:, 2:4] , iris.target)
print(classifier2.score)
predicted2 = classifier2.predict(test_data)
print(predicted2)

plt.scatter(xx, yy , c=predicted1)
plt.title("MLPClassifier")

x_ind = 2
y_ind = 3
# plt.scatter(iris.data[:, x_ind], iris.data[:, y_ind], c=iris.target)
plt.scatter(iris.data[:50, x_ind], iris.data[0:50, y_ind])
plt.scatter(iris.data[50:100, x_ind], iris.data[50:100, y_ind])
plt.scatter(iris.data[100:150, x_ind], iris.data[100:150, y_ind])
plt.xlabel(iris.feature_names[x_ind])
plt.ylabel(iris.feature_names[y_ind])


plt.scatter(xx, yy , c=predicted2)
plt.title("Naive Bayesian Classifier")

x_ind = 2
y_ind = 3
# plt.scatter(iris.data[:, x_ind], iris.data[:, y_ind], c=iris.target)
plt.scatter(iris.data[:50, x_ind], iris.data[0:50, y_ind])
plt.scatter(iris.data[50:100, x_ind], iris.data[50:100, y_ind])
plt.scatter(iris.data[100:150, x_ind], iris.data[100:150, y_ind])
plt.xlabel(iris.feature_names[x_ind])
plt.ylabel(iris.feature_names[y_ind])


plt.show()
```
