#While loop
    > in python the condition in while loop can have a else statement 

```
i = 0
while i<5 :
    print(i ,"in while loop")
    i += 1 
else:
    print("(optional) while loop finished")



# break, continue 
i = 0
while i<10 :
    i += 1
    if i == 8:
        break
    if i%2 == 0:
        continue
    print(i ,"in while loop")
```
