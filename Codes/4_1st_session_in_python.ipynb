```
#list
new_list=[] # empty list
new_list= [1,14,5,"string",6.55 , 20+9j, 50, "befor last", 234 , 0.05]
### LIST IS A MUTABLE OBJECT, meaning the type of the list is changable

print(len(new_list))


### list methods
new_list.append("new item")
print(new_list, type(new_list))
print(len(new_list))

test = "string" in new_list
print(test)
new_list.remove("string")
print(len(new_list))

test = "string" in new_list ## can be done for tuple,string
print(test)



### list slicing >>> to work with a portion of a list
start = 2
end = 6
sliced_list= new_list[ start : end] # from the start index to the (end-1) index
print("sliced list= " ,sliced_list)

step = 3
sliced_list= new_list[ start : end : step]
print("sliced list= " ,sliced_list)

### negative indexing
sliced_list = new_list[-5 : -1]
print(sliced_list)

### list sortig, all elements must be of the same type
int_list = [1,14,5,6.55, 50, 234 , 0.05]

### list assigning>> uses pointer to address
int_list2 = int_list

int_list.append(0)

print("\n",int_list, int_list2)

#### to avoid this we must 
int_list2 = int_list.copy()
int_list.pop(0)
print("\n",int_list, int_list2)
```

```
#### set 
a = {2,3,5,1,4,2,2,2,2,6,6,8,4}
b = {10,10,10,69,8,7}
print(a)

a = a.union(b)
print(a)

if a.isdisjoint(b):
    print("Disjoint, has no common")
else: print("has common")

a.remove(69) ### remove() does not check if the value exists or not, so it may generate error
print(a)
a.discard(69) ### discard() checks if the value exists, and removes, so does not generate error 
print(a)

a.clear()
print(a)
```

```
class Triangle:
    a = 0.0
    b = 0.0
    c = 0.0
    def __init__(self, a, b, c):
        self.a = float(a)
        self.b = float(b)
        self.c = float(c)

    def Area (self):
        p = (self.a + self.b + self.c)/ 2
        return  ( p * (p-self.a) * (p-self.b) * (p-self.c) )**0.5


c = Triangle(5,6,7)
print(c.Area())
```
