```
from gensim.models import Word2Vec
from sklearn.decomposition import PCA
from matplotlib import pyplot
```

```
#training data
sentences = [['this', 'is', 'the', 'first', 'sentence', 'for', 'word2vec'],
			['this', 'is', 'the', 'second', 'sentence'],
			['yet', 'another', 'sentence'],
			['one', 'more', 'sentence'],
			['and', 'the', 'final', 'sentence']]

# model.wv.save_word2vec_format("model.txt", binary=False)
# model.wv.save_word2vec_format("model.bin")
model.save("model.bin")


model = Word2Vec.load("/content/model.bin")
words = list(model.wv.vocab)
print(words)

print("\n sentence = \n",model["sentence"], "\n")
```

##Visualize Word encoding

```
X = model[model.wv.vocab]

pca = PCA(n_components=2) #for 2D representation
x_pca = pca.fit_transform(X)

print(x_pca)

pyplot.scatter(x_pca[: , 0], x_pca[: ,1])
for i, word in enumerate(words):
    pyplot.annotate(word, xy=(x_pca[i, 0], x_pca[i, 1]) )
pyplot.show()
```
