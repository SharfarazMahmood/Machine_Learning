```
from sklearn.datasets  import load_iris
import matplotlib.pyplot as plt
```

#Scikit learn

##Load IRIS dataset from sklearn


*   check samples, features
*   check target/ classes





```
iris = load_iris()
print(type(iris), "\n ",type(iris.data))
# print(iris.data)

n_sample, n_features = iris.data.shape
print(n_sample, n_features)
print(iris.target)
print(iris.target_names)


## slicing
print(iris.data[:10])
```

##Plot some data

```
plt.subplot()
plt.scatter(iris.data[:, 0], iris.data[:, 1], c=iris.target)
plt.xlabel(iris.feature_names[2])
plt.ylabel(iris.feature_names[3])


plt.subplot()
plt.scatter(iris.data[:, 2], iris.data[:, 3], c=iris.target)
plt.xlabel(iris.feature_names[2])
plt.ylabel(iris.feature_names[3])
plt.show()
```

#Use a built in NN Model to classify


## MLPClassifier

https://scikit-learn.org/stable/modules/generated/sklearn.neural_network.MLPClassifier.html#sklearn.neural_network.MLPClassifier

> 1 hidden layer, with 5 neuron, 



```
from sklearn.neural_network import MLPClassifier

mlp_model = MLPClassifier(hidden_layer_sizes=(5), activation="logistic", max_iter=1000, learning_rate_init=.01 )
print(type(mlp_model))


## input data in the model And train the model
mlp_model.fit(iris.data, iris.target)

## predict
output_classes = mlp_model.predict(iris.data)
print(output_classes)

### predict a single data/instance
test = (5.1, 3.5, 1.4, 0.2)
predicted_class_single_data = mlp_model.predict( [test] )
print(predicted_class_single_data, iris.target_names[predicted_class_single_data])


## class probability 
predicted_probability = mlp_model.predict_proba([test])
print(predicted_probability)
```

##sklearn metrics: model report
https://scikit-learn.org/stable/modules/generated/sklearn.neural_network.MLPClassifier.html#sklearn.neural_network.MLPClassifier

```
from sklearn import metrics as MLP_metrics

## confusion matrix
conf_mat = MLP_metrics.confusion_matrix(iris.target,  output_classes)
print(conf_mat)

## report 
report = MLP_metrics.classification_report(iris.target, output_classes)
print(report)

```

```
from sklearn.datasets import load_iris
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics 

iris = load_iris()
data = iris.data
target = iris.target
x_train, x_test, y_train, y_test = train_test_split(data, target, test_size=0.20, shuffle=True,random_state=1 )
print( len(y_train), len(y_test) )

model = MLPClassifier(hidden_layer_sizes=(100,), activation="relu", solver="sgd",
                      learning_rate_init=0.001, max_iter=500, shuffle=True, tol=.0001, 
                      n_iter_no_change=5 )
model.fit(x_train, y_train)
test_prediction = model.predict(x_test)

conf_mat = metrics.confusion_matrix(y_test, test_prediction)
# print( conf_mat )

cohen_kappa = metrics.cohen_kappa_score( y_test , test_prediction )
print("cohen_kappa score:", cohen_kappa)
# import seaborn as sns
# sns.heatmap( conf_mat )

print(y_test)
print(test_prediction)
```
