```
#open/load the file
try:
    f = open("hamlet.txt")
    hamlet = f.read()
    f.close()
except Exception:
    print("file error")

# examin the file
# print(type(hamlet), len(hamlet))

# let's split the text to words
#   `default split = splits the text without the whitespaces 
hamlet_words = hamlet.split()
# print(type(hamlet_words), len(hamlet_words))
# print(hamlet_words[:10])

### frequency of the words 
#  convert the case of the characters, upper to lower or vice versa 
hamlet_words_lower = [w.lower() for w in hamlet_words] 
print(hamlet_words_lower[:10])

word_count = hamlet_words_lower.count("hamlet")
# print(word_count)

### fid the unique words of the text 
unique_words = set(hamlet_words_lower)
print(len(unique_words), type(unique_words))
print(unique_words)

#### make a ""dictionary"" for the unique words with theies corressponding frequencies
## unique word
# key = word
# value = number of occurance 
word_frequency = {}

for word in unique_words:
    word_frequency[word] =  hamlet_words_lower.count(word)
print(len(word_frequency))
```
