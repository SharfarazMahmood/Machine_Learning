```
# mount google drive to access files, skip this part if drive is not needed
from google.colab import drive
drive.mount('/content/drive')
%cd /content/drive/My Drive/Work_FTFL_GIGATECH

## read data from drive
import pandas as pd

data_xlxs = pd.read_excel(open('/content/drive/My Drive/Work_FTFL_GIGATECH/facebook.xlsx', 'rb'),
                          sheet_name='Sheet1')
data_df = pd.DataFrame(data_xlxs, columns=data_xlxs.columns[0:])

# print(data_df.columns[0:])
text_df = data_df.iloc[0:, 0:1]

print(text_df.shape)
```

```
### read the data from dataframe and split them into words
accepting_symbols= 'ঀঁং ঃঅআইঈউঊঋঌএঐওঔকখগঘঙচছজঝঞটঠডঢণতথদধনপফবভমযরলশষসহড়ঢ়য়ৠৰৱ ন্ত ন্য দ্ব ন্ধ ়  হ্রদ  সম্বন্ধে ম্ব পৃ  া ন্তি   ি  ী  ু  ূূ  ৃ	ৄ ে	ৈ ো  ৌ 	্ ৎ  ৗ ৡ  	ৢ	ৣ ০১২৩৪৫৬৭৮৯  ৲ ৳	৴ ৵	৶ ৷	৸৹৺৻ !,'

text_words = []
text_words_set = []

for index, row in text_df.iterrows():
    # print(index)
    words = list( row['Text'].split() )
    # print(len(words))
    for i in range(0, len(words)):
        # print(i)
        words[i]= ''.join(c for c in words[i] if c in accepting_symbols)
        if len(words[i])==0:
            words.pop()
    text_words.append(words)
    text_words_set.append(set(words))
        
    # print(words)
```

```
# find similarity between the data instances
similar_text_ind = []
text_to_remove_ind = []
threshold_score = 0.25
for i in range (0 , len(text_words)):
    # print()
    for j in range (i+1 , len(text_words)):

        jac_scr = len(set(text_words[i]).intersection(set(text_words[j]))) / len(set(text_words[i]).union(set(text_words[j])))
        # print("index:",i , " -- ", j," JAC score:", jac_scr)

        if jac_scr > threshold_score >= 0.10:
            similar_text_ind.append([i, j])
            text_to_remove_ind.append(j)
print(similar_text_ind)
text_to_remove_ind = set( text_to_remove_ind )
print(text_to_remove_ind)

# for item in similar_text_ind:
#     print(text_words[item[0]])
#     print(text_words[item[1]])
#     print(  set(text_words[item[0]]).intersection(set(text_words[item[1]])) )
```
