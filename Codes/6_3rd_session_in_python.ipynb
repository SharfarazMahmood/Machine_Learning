```
try:
    fw = open("write_test.txt","a")

    dict_of_object = { "Jamee":[1, "jan",2000] , "Sharfaraz":[31, "jun",1990], "Mahmood":[9, "feb",1995]}
    for person in dict_of_object:
        value = dict_of_object[person]
        day = value[0]
        month = value[1]
        year = value[2]
        output= "{} was born in {} {} in {}\n".format(person, day, month, year)

        fw.writelines(output)
    fw.close()
except FileExistsError as ferror:
    print(ferror)
except Exception as e:
    print(e)
finally:
    print("Done")
```

```
class Car:
    make = None
    model = None
    wheels = None
    speed = None

    def __init__(self, make, model, wheels):
        self.make = make
        self.model = model
        self.wheels = wheels
        self.speed=0

    def accelerate (self):
        self.speed += 10
    def brake (self):
        self.speed -= 10
    def getModel (self):
        return self.model
    def getSpeed (self):
        return self.speed



car1 = Car("BMW", "Model_4", 4)
print("This is a",car1.make, car1.getModel(),"with", car1.wheels, "wheels")
car1.accelerate()
car1.accelerate()
car1.accelerate()
print(car1.getSpeed())
car1.brake()
car1.brake()
print(car1.getSpeed())
```
