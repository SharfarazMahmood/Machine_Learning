```
# digits datasets

from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split

digits = load_digits()

print(digits.data.shape, " number of datas, number of features")

n=4
print(digits.data[n])

print(digits.target_names)
plt.gray()
plt.imshow(digits.data[n].reshape(8,8))
plt.show()


x_train, x_test, y_train, y_test = train_test_split(digits.data, digits.target , test_size = 0.30, random_state = 1)
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```
#classify the digits dataset with MLP

from sklearn.neural_network import MLPClassifier

from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

clf = MLPClassifier()
clf.fit(x_train, y_train)

y_predicted =  clf.predict(x_test)

# print(y_predicted)

conf_mat = confusion_matrix(y_test, y_predicted)

print(conf_mat)


report = classification_report(y_test, y_predicted)
print(report)
```
